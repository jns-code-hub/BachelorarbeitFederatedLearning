```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
def sgd_averaged_acc():
    file_path_sgd_42 = f"results/42/SGD.json"
    with open(file_path_sgd_42) as json_file:
        sgd_42_data = json.load(json_file)

    file_path_sgd_51588 = f"results/51588/SGD.json" 
    with open(file_path_sgd_51588) as json_file:
        sgd_51588_data = json.load(json_file)

    file_path_sgd_508 = f"results/508/SGD.json" 
    with open(file_path_sgd_508) as json_file:
        sgd_508_data = json.load(json_file)

    t = np.array([
        sgd_42_data["sgd_baseline"]["val_acc"], 
        sgd_508_data["sgd_baseline"]["val_acc"], 
        sgd_51588_data["sgd_baseline"]["val_acc"]
    ])

    avg_acc = np.mean(t, axis=0)
    std = np.std(t, axis=0)

    return avg_acc, std

def sgd_averaged_loss():
    file_path_sgd_42 = f"results/42/SGD.json"
    with open(file_path_sgd_42) as json_file:
        sgd_42_data = json.load(json_file)

    file_path_sgd_51588 = f"results/51588/SGD.json" 
    with open(file_path_sgd_51588) as json_file:
        sgd_51588_data = json.load(json_file)

    file_path_sgd_508 = f"results/508/SGD.json" 
    with open(file_path_sgd_508) as json_file:
        sgd_508_data = json.load(json_file)

    t = np.array([
        sgd_42_data["sgd_baseline"]["val_loss"], 
        sgd_508_data["sgd_baseline"]["val_loss"], 
        sgd_51588_data["sgd_baseline"]["val_loss"]
    ])

    avg_loss = np.mean(t, axis=0)
    std = np.std(t, axis=0)

    return avg_loss, std
```

```python
def average_Fed_acc(aggregation,label, quantity, mu, beta):
    path_42 = f"results/42/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_42) as json_file:
        data_42 = json.load(json_file)

    path_508 = f"results/508/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_508) as json_file:
        data_508 = json.load(json_file)

    path_51588 = f"results/51588/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_51588) as json_file:
        data_51588 = json.load(json_file)

    t = np.array([
        data_42["federated"]["global_val_acc"],
        data_508["federated"]["global_val_acc"],
        data_51588["federated"]["global_val_acc"]
    ])

    avg_acc = np.mean(t, axis=0)
    std = np.std(t, axis=0)

    return avg_acc, std

def average_Fed_loss(aggregation,label, quantity, mu, beta):
    path_42 = f"results/42/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_42) as json_file:
        data_42 = json.load(json_file)

    path_508 = f"results/508/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_508) as json_file:
        data_508 = json.load(json_file)

    path_51588 = f"results/51588/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_51588) as json_file:
        data_51588 = json.load(json_file)

    t = np.array([
        data_42["federated"]["global_val_loss"],
        data_508["federated"]["global_val_loss"],
        data_51588["federated"]["global_val_loss"]
    ])

    avg_loss = np.mean(t, axis=0)
    std = np.std(t, axis=0)

    return avg_loss, std

def average_Fed_layer_weight_divergence(aggregation,label, quantity, mu, beta):
    path_42 = f"results/42/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_42) as json_file:
        data_42 = json.load(json_file)

    path_508 = f"results/508/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_508) as json_file:
        data_508 = json.load(json_file)

    path_51588 = f"results/51588/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(path_51588) as json_file:
        data_51588 = json.load(json_file)

    t = np.array([
        data_42["federated"]["layer_weight_divergence"],
        data_508["federated"]["layer_weight_divergence"],
        data_51588["federated"]["layer_weight_divergence"]
    ])
    
    layers_to_avg = [0,2,4,6]
    t_filtered_layers = t[:, :, layers_to_avg]

    avg_layer_wd = np.mean(t_filtered_layers, axis=0)
    std = np.std(t_filtered_layers, axis=0)

    return avg_layer_wd, std

    
```

```python
def plot_FedAvg_Label_Skew_Accuracy(ax):
    sgd_baseline_acc, sgd_acc_std = sgd_averaged_acc()
    iid_acc, iid_loss_std = average_Fed_acc(0, 1000000, 1000000, 0, 0)
    strong_label_skew_acc, strong_label_skew_acc_std = average_Fed_acc(0, 0.001, 1000000, 0, 0)
    mild_label_skew_acc, mild_label_skew_acc_std = average_Fed_acc(0, 0.1, 1000000, 0, 0)
    light_label_skew_acc, light_label_skew_acc_std = average_Fed_acc(0, 1, 1000000, 0, 0)

    print("Avg Acc Strong Label Skew", np.mean(strong_label_skew_acc[-20:]))
    print("Letzte Acc Strong Label Skew", strong_label_skew_acc[-1])
    print("Avg Acc Mild Label Skew", np.mean(mild_label_skew_acc[-20:]))
    print("Letzte Acc Mild Label Skew", mild_label_skew_acc[-1])
    print("Avg Acc Light Label Skew", np.mean(light_label_skew_acc[-20:]))
    print("Letzte Acc Light Label Skew", light_label_skew_acc[-1])
    print("Avg Acc IID", np.mean(iid_acc[-20:]))
    print("Letzte Acc IID", iid_acc[-1])
    print("Avg Acc SGD", np.mean(sgd_baseline_acc[-20:]))
    print("Letzte Acc SGD", sgd_baseline_acc[-1])

    ax.plot(sgd_baseline_acc, label="SGD", color="black")
    ax.plot(iid_acc, label="IID", linestyle="--", color="gray")
    
    ax.plot(strong_label_skew_acc, label="Starker Label Skew", color="red")
    ax.fill_between(range(300), strong_label_skew_acc - strong_label_skew_acc_std, strong_label_skew_acc + strong_label_skew_acc_std, alpha=0.08, color="red")
    
    ax.plot(mild_label_skew_acc, label="Mittlerer Label Skew", color="orange")
    ax.fill_between(range(300), mild_label_skew_acc - mild_label_skew_acc_std, mild_label_skew_acc + mild_label_skew_acc_std, alpha=0.08, color="orange")

    ax.plot(light_label_skew_acc, label="Leichter Label Skew", color="green")
    ax.fill_between(range(300), light_label_skew_acc - light_label_skew_acc_std, light_label_skew_acc + light_label_skew_acc_std, alpha=0.08, color="green")
    
    ax.set_title("Label Skew Accuracy")
    ax.set_xlabel("Runden")
    ax.set_ylabel("Accuracy")

    ax.grid(True)

def plot_FedAvg_Label_Skew_Loss(ax):
    sgd_baseline_loss, sgd_loss_std = sgd_averaged_loss()
    iid_loss, iid_loss_std = average_Fed_loss(0, 1000000, 1000000, 0, 0)
    strong_label_skew_loss, strong_label_skew_loss_std = average_Fed_loss(0, 0.001, 1000000, 0, 0)
    mild_label_skew_loss, mild_label_skew_loss_std = average_Fed_loss(0, 0.1, 1000000, 0, 0)
    light_label_skew_loss, light_label_skew_loss_std = average_Fed_loss(0, 1, 1000000, 0, 0)

    ax.plot(sgd_baseline_loss, label="SGD", color="black")
    ax.plot(iid_loss, label="IID", linestyle="--", color="gray")
    
    ax.plot(strong_label_skew_loss, label="Starker Label Skew", color="red")
    ax.fill_between(range(300), strong_label_skew_loss - strong_label_skew_loss_std, strong_label_skew_loss + strong_label_skew_loss_std, alpha=0.08, color="red")
    
    ax.plot(mild_label_skew_loss, label="Mittlerer Label Skew", color="orange")
    ax.fill_between(range(300), mild_label_skew_loss - mild_label_skew_loss_std, mild_label_skew_loss + mild_label_skew_loss_std, alpha=0.08, color="orange")

    ax.plot(light_label_skew_loss, label="Leichter Label Skew", color="green")
    ax.fill_between(range(300), light_label_skew_loss - light_label_skew_loss_std, light_label_skew_loss + light_label_skew_loss_std, alpha=0.08, color="green")
    
    ax.set_title("Label Skew Loss")
    ax.set_xlabel("Runden")
    ax.set_ylabel("Loss")
    ax.legend()
    ax.grid(True)


def create_combined_plot():
    plt.rc('font', size=12)
    plt.rc('axes', titlesize=16)
    plt.rc('axes', labelsize=14)
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    plot_FedAvg_Label_Skew_Accuracy(axes[0])  
    plot_FedAvg_Label_Skew_Loss(axes[1])     

    plt.tight_layout()
    fig.savefig("fedavg_label_skew.pdf", bbox_inches="tight")
    plt.show()    

create_combined_plot()
```

```python
def load_files(seed, aggregation, label, quantity, mu, beta):
    file_path = f"results/{seed}/{aggregation}_Label_{label}_Quantity_{quantity}_MU{mu}_Beta_{beta}/results.json"
    with open(file_path) as json_file:
        data = json.load(json_file)

    return data
```

```python
def calc_mssd(arr):
    masd = np.mean(np.diff(arr)**2)
    return masd
```
